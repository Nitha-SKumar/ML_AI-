#Lowercasing

```python
df["reviews.title"] = df["reviews.title"].astype(str)
df["reviews.text"] = df["reviews.text"].astype(str)
```

```python
df["reviews.title"] = df["reviews.title"].str.lower()
df["reviews.text"] = df["reviews.text"].str.lower()
df.head()
```

#Removing Punctuation

```python
import string
string.punctuation
def remove_punct(text):
    text_nopunct = "".join([char for char in text if char not in string.punctuation])
    return text_nopunct


df["reviews.title_clean"] = df["reviews.title"].apply(lambda x: remove_punct(x))
df["reviews.text_clean"] = df["reviews.text"].apply(lambda x: remove_punct(x))

df.head()
```

#Tokenization

```python
import re

def tokenize(text):
    tokens = re.split('\W+', text)
    return tokens

df['reviews.title_tokenized'] = df['reviews.title_clean'].apply(lambda x: tokenize(x.lower()))

df['reviews.text_tokenized'] = df['reviews.text_clean'].apply(lambda x: tokenize(x.lower()))

df.head()
```


#Remove Stopwords

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
```

```python
STOPWORDS = set(stopwords.words('english'))
def remove_stopwords(text):
    """custom function to remove the stopwords"""
    return " ".join([word for word in str(text).split() if word not in STOPWORDS])


df['reviews.title_stopwords'] = df['reviews.title_clean'].apply(lambda text: remove_stopwords(text))
df['reviews.text_stopwords'] = df['reviews.text_clean'].apply(lambda text: remove_stopwords(text))
df.head()
```

#Removal of Frequent words

```python
from collections import Counter
cnt = Counter()
for text in df["reviews.title_stopwords"].values:
    for word in text.split():
        cnt[word] += 1


cnt.most_common(10)
```

```python
from collections import Counter
cnt = Counter()
for text in df["reviews.text_stopwords"].values:
    for word in text.split():
        cnt[word] += 1


cnt.most_common(10)
```

#Stemming

```python
from nltk.stem.porter import PorterStemmer

# Drop the two columns
#df.drop(["text_wo_stopfreq", "text_wo_stopfreqrare"], axis=1, inplace=True)

stemmer = PorterStemmer()
def stem_words(text):
    return " ".join([stemmer.stem(word) for word in text.split()])

df["reviews.text_stemmed"] = df["reviews.text_stopwords"].apply(lambda text: stem_words(text))
df["reviews.title_stemmed"] = df["reviews.title_stopwords"].apply(lambda text: stem_words(text))
df.head()
```

#Lemmatization

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])



df["reviews.text_lemmaatized"] = df["reviews.text_stopwords"].apply(lambda text: lemmatize_words(text))
df["reviews.title_lemmatized"] = df["reviews.title_stopwords"].apply(lambda text: lemmatize_words(text))
df.head()
```

```python
# Drop the two columns which are no more needed
df.drop(["reviews.title_clean","reviews.text_clean",	"reviews.title_tokenized",	"reviews.text_tokenized",	"reviews.title_stopwords",	"reviews.text_stopwords"], axis=1, inplace=True)
```
