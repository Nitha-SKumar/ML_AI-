```python
import pandas as pd
import numpy as numpy
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters = 5, n_init = 20)
```

```python
km.fit(data)
```

```python
cluster_labels = km.predict(data)
```

```python
plt.scatter(data['1'], data['2'], c = cluster_labels )

plt.show()
```

```python
km.inertia_
```

```python
inertia_list = []


for i in range(1,11):
  km = KMeans(n_clusters = i, n_init = 20)
  km.fit(data)
  inertia = km.inertia_
  inertia_list.append(inertia)
```

```python
inertia_list
```

```python
plt.plot(range(1,11), inertia_list)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.show();
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
mergings = linkage(x, method = "ward")
```

```python
plt.figure(figsize= (20, 6))
dendrogram(mergings)
plt.axhline(190, color = "red", linestyle = "--")
plt.show()
```

```python
cluster_labels = fcluster(mergings, 190, criterion = "distance")
```

```python
pd.Series(cluster_labels).value_counts()
```

```python

plt.scatter(pc_data['pc1'], pc_data['pc2'], c = cluster_labels )
```
